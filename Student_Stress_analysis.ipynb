```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#For ignoring warning
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['Kindly Rate your Sleep Quality üò¥']=le.fit_transform(df['Kindly Rate your Sleep Quality üò¥'])
df['How many times a week do you suffer headaches ü§ï?']=le.fit_transform(df['How many times a week do you suffer headaches ü§ï?'])
df['How would you rate you academic performance üë©‚Äçüéì?']=le.fit_transform(df['How would you rate you academic performance üë©‚Äçüéì?'])
df['how would you rate your study load?']=le.fit_transform(df['how would you rate your study load?'])
df['How many times a week you practice extracurricular activities üéæ?']=le.fit_transform(df['How many times a week you practice extracurricular activities üéæ?'])
df['How would you rate your stress levels?']=le.fit_transform(df['How would you rate your stress levels?'])
```

**CORRELATION**

```python
#Finding Correlation
cn=df.corr()
cn
```

```python
#Correlation
cmap=sns.diverging_palette(260,-10,s=50, l=75, n=6,
as_cmap=True)
plt.subplots(figsize=(18,18))
sns.heatmap(cn,cmap=cmap,annot=True, square=True)
plt.show()
```

```python
kot = cn[cn>=.40]
plt.figure(figsize=(12,8))
sns.heatmap(kot, cmap="Blues")
```
